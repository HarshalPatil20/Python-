```python
import matplotlib.pyplot as plt
```

line graph

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y)
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("y=x**2")
```

bar graph

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.bar(x,y,color='Red')
plt.xlabel("x value")
plt.ylabel("y value")
```

##Horizontal bar graph

```python
plt.barh(x,y,color='Black')
```

#scater plot


```python
plt.scatter(x,y,color='Blue',alpha=0.9,s=300,)
```

#Histogram


```python
plt.hist(y)
```

#pie chart

```python
plt.pie(x,labels=x)
```
